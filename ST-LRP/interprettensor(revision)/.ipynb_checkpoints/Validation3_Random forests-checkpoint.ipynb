```python
from sklearn.ensemble import RandomForestClassifier
import pickle
import copy
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib
import tensorflow as tf
import pdb
import scipy.io as sio
import numpy as np
import pandas as pd
from pylab import rcParams
import math
import time

##Input Data

F1=open("E:/holiday_2_train_2016",'rb')
#F1=open("E:/revisiondata/trainnorm_2016",'rb')
train=pickle.load(F1)
F1.close()

F2=open("E:/holiday_2_test_2016",'rb')
#F2=open("E:/revisiondata/testnorm_2016",'rb')
test=pickle.load(F2)
F2.close()

F4=open("E:/holiday_2_test_2017",'rb')
#F4=open("E:/revisiondata/testnorm_2017",'rb')
data2017=pickle.load(F4)
F4.close()
```

```python
##Data reshape

X1=[]
Y1=[]
for i in range(len(train)):
    if(train[i][1][0]==1):
        Y1.append(0)
    else:
        Y1.append(1)
    #tmp=copy.deepcopy(np.reshape(np.array(train[i][0]),(30*30*48)))
    tmp=copy.deepcopy(np.sum(np.reshape(np.array(train[i][0]),(900,48)),axis=0))
    #tmp=copy.deepcopy(np.sum(np.reshape(np.array(train[i][0]),(900,48)),axis=1))
    X1.append(tmp)

X2=[]
Y2=[]
for i in range(len(test)):
    if(test[i][1][0]==1):
        Y2.append(0)
    else:
        Y2.append(1)
    #tmp=copy.deepcopy(np.reshape(np.array(test[i][0]),(30*30*48)))
    tmp=copy.deepcopy(np.sum(np.reshape(np.array(test[i][0]),(900,48)),axis=0))
    #tmp=copy.deepcopy(np.sum(np.reshape(np.array(test[i][0]),(900,48)),axis=1))
    X2.append(tmp)



X4=[]
Y4=[]
for i in range(len(data2017)):
    if(data2017[i][1][0]==1):
        Y4.append(0)
    else:
        Y4.append(1)
    #tmp=copy.deepcopy(np.reshape(np.array(data2017[i][0]),(30*30*48)))
    tmp=copy.deepcopy(np.sum(np.reshape(np.array(data2017[i][0]),(900,48)),axis=0))
    #tmp=copy.deepcopy(np.sum(np.reshape(np.array(data2017[i][0]),(900,48)),axis=1))
    X4.append(tmp)
```

```python
##Validation3, Random forests

importance=[]
accuracy=[]
repeat=50
for n in range(repeat):
    clf=RandomForestClassifier(n_estimators=10).fit(X1,Y1)
    tmp_i=clf.feature_importances_
    importance.append(tmp_i)
    
    tmp_p=[]
    result=clf.predict(X2)
    p=0.0
    for i in range(len(result)):
        if(result[i]==Y2[i]):
            p+=1
    p=p/len(result)
    tmp_p.append(p)

    result=clf.predict(X1)
    p=0.0
    for i in range(len(result)):
        if(result[i]==Y1[i]):
            p+=1
    p=p/len(result)
    tmp_p.append(p)


    result=clf.predict(X4)
    p=0.0
    for i in range(len(result)):
        if(result[i]==Y4[i]):
            p+=1
    p=p/len(result)
    tmp_p.append(p)
    
    accuracy.append(tmp_p)
```

```python
##Get the classification accuracy and element importance

result1=np.mean(np.array(accuracy),axis=0)
for i in result1:
    print(str(i))
print('-------------------')
result2=np.mean(np.array(importance),axis=0)
for i in result2:
    print(str(i)+',')
```
